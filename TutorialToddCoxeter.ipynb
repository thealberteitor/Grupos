# Algoritmo de Todd Coxeter

Sea G un grupo definido por una presentación \\( G = \langle X \mid R \rangle \\), donde \\( X \\) es el conjunto de generadores y \\( R\\) el conjunto de relatores. Sea \\(H = \langle h_1, h_2,...,h_r \rangle \leq G \\), donde los generadores \\( h_i \\) son palabras en el alfabeto \\( X^{\pm 1} \\).






```python
from Group import *
from ToddCoxeter import CosetTable, readGroup
```

\\[ G = \langle a, b, c \mid a^2, b^2, c^3, aba^{-1}b^{-1}, cac^{-1}b^{-1}a^{-1}, cbc^{-1}a^{-1}    \rangle ,\\]

\\[ H = \langle b \rangle .\\]

```python
f = readGroup("Groups/A4.txt")
f
```

```python
C = CosetTable(f)
C.CosetEnumeration()
print(C.coset_table())
```

```python
C.schreier_graph(notes=False)
```

```python
def print_gens(gens):
    for i in range(len(gens)):
        print(f"g{i} = {gens[i]}")
        
generators = C.getGenerators()
print_gens(generators)
```

```python
G = Group(elems=generators)
print(G)
```

```python
G.table()
```

```python
G.is_abelian()
```

\\[ G \cong D_6 \\]
\\[ G \cong Q_3  \\]
\\[ G \cong A_4  \\]



```python
D6 = DihedralGroup(6)
Q3 = QuaternionGroupGeneralised(3)
A4 = AlternatingGroup(4)

G.is_isomorphic(D6), G.is_isomorphic(Q3), G.is_isomorphic(A4)
```
